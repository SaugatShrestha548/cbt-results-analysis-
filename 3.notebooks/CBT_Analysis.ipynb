```python
import numpy as np
import pandas as pn
import matplotlib.pyplot as plt
import seaborn as sb
from scipy import stats 
```

```python
db = pn.read_excel(r"C:\Users\BROTHER COMPUTER\Desktop\pdf_merged\ds.xlsx")
```

```python
mn = np.mean(db["CBT Score"])
```

```python
sb.histplot(x = "CBT Score", data = db)
plt.plot([mn for i in range(0,15)] ,[i for i in range(0,15)],c = 'red',linestyle = "--", label = f'Mean:{mn :.2f}')
plt.plot([md for i in range(0,15)] ,[i for i in range(0,15)],c = 'green',linestyle = "dashdot", label = f'Median:{md :.2f}')
plt.axvline(x = mo.mode, color = 'orange',linestyle = "solid", label = f'Mode:{mo.mode :.2f}')
plt.legend()
plt.show()
```

```python
md = np.median(db["CBT Score"])
```

```python
mo = stats.mode(db["CBT Score"])
```

```python
sb.kdeplot(x = "CBT Score", data = db)
plt.axvline(x = mn, color ='red', linestyle = "--", label = f'Mean:{mn :.2f}')
plt.axvline(x = md, color ='green',linestyle = "dashdot", label = f'Median:{md :.2f}')
plt.axvline(x = mo.mode, color = 'orange',linestyle = (0, (8, 2, 8, 5,1,2)), label = f'Mode:{mo.mode :.2f}')
plt.legend()
```
